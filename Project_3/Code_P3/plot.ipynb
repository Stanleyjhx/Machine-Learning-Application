This notebook plots all the data we have, please check result folder for the text files we're using

```python
import matplotlib.pyplot as plt
import numpy as np
import xlwt
import os

with open('mlp0_test.txt','r+') as fd:
  e = list(range(1,101))
  p = []
  #read all accuracy from the performance report
  for text in fd.readlines():
    y=text.split(':')[3].strip("\n").strip('%').strip()
    p.append(y.split("\n"))

  p = list(np.concatenate(p).flat)
  p = [round(float(i)/100,3) for i in p]
  print(p)
  print(e)

fd.close()

with open('mlp0_train.txt','r+') as fd2:
  e = list(range(1,101))
  p_train = []
  #read all accuracy from the performance report
  for text in fd2.readlines():
    y=text.split(':')[3].strip("\n").strip('%').strip()
    p_train.append(y.split("\n"))

  
  p_train = list(np.concatenate(p_train).flat)
  p_train = [round(float(i)/100,3) for i in p_train]
  print(p_train)
  print(e)

fd2.close()

plt.plot(e, p, color='green',  linewidth = 2, label="test")
plt.plot(e, p_train, color='orange',  linewidth = 2, label="train")
plt.legend(loc="lower right")
  
# setting x and y axis range
plt.ylim(0,1)
plt.xlim(1,100)
  
# naming the x axis
plt.xlabel('Iterations')
# naming the y axis
plt.ylabel('Accuracy')
  
# giving a title to my graph
plt.title('Zero hidden layer MLP test with training & testing set performance')
  
# function to show the plot
plt.show()
```
